```python
import zipfile
import os
import tensorflow as tf

# Unzip the folder
zip_path = 'model_6_savedmodel.zip'
extract_path = 'model_6_savedmodel'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Load the model
loaded_model = tf.keras.models.load_model(extract_path)

# Now you can use the loaded_model
```

```python
loaded_model_6 = tf.keras.models.load_model("model_6_savedmodel")
```

```python
# Assuming you have a sentence
new_sentence = "This is a new sentence for prediction."

# Use the loaded model to make predictions
pred_prob = loaded_model_6.predict([new_sentence])
pred_label = tf.squeeze(tf.round(pred_prob)).numpy()

# Print the prediction
print(f"Predicted Label: {int(pred_label)}")
print(f"Prediction Probability: {pred_prob[0][0]}")
```

```python
# Read the text file with sentences
file_path = "result.txt"
with open(file_path, "r", encoding="utf-8") as file:
    sentences = file.readlines()

# Use the loaded model to make predictions for each sentence
for sentence in sentences:
    # Clean the sentence if needed
    cleaned_sentence = sentence.strip()  # You may need additional cleaning based on your data

    # Make predictions
    pred_prob = loaded_model_6.predict([cleaned_sentence])
    pred_label = int(tf.squeeze(tf.round(pred_prob)).numpy())

    # Print the prediction for each sentence
    print(f"Sentence: {cleaned_sentence}")
    print(f"Predicted Label: {pred_label} (Probability: {pred_prob[0][0]:.4f})")
    print("-" * 30)
```

```python
# Assuming you have a text file named "your_text_file.txt" with one sentence per line
with open("result.txt", "r") as file:
    for line in file:
        # Use the loaded model to make predictions
        pred_prob = loaded_model_6.predict([line])
        pred_label = tf.squeeze(tf.round(pred_prob)).numpy()

        # Print the prediction if the label is 1
        if pred_label == 1:
            print("Predicted Label: 1")
            print(f"Sentence: {line.strip()}")
            print(f"Prediction Probability: {pred_prob[0][0]}")
            print("------------------------------")
```

```python
# Assuming you have a text file named "your_text_file.txt" with one sentence per line
with open("result.txt", "r") as file:
    for line in file:
        # Use the loaded model to make predictions without printing progress
        pred_prob = loaded_model_6.predict([line], verbose=0)
        pred_label = tf.squeeze(tf.round(pred_prob)).numpy()

        # Print the prediction if the label is 1
        if pred_label == 1:
            print("Predicted Label: 1")
            print(f"Sentence: {line.strip()}")
            print(f"Prediction Probability: {pred_prob[0][0]}")
            print("------------------------------")
```
